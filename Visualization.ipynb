# Types
1. Bar charts
* Uses bars to compare discrete categories of data. The bars can either be horizontal or vertical

Best for:
  * Comparing values across categories
  * Displaying part_to_whole relationships in data

eg. sales per region, population by age group

2. Line chart
* Plots data points along a continous line, often used to show trend over time.

Best for:
 * Time series analysis(Stock prices)
 * Showing trends, patterns and changes or intervals

3. Pie charts
* Circular chart divided into slices to show proportions of a whole.

Best for:
 * Showing percentages of proportional data fora  small number of categories[max=5]

 4. Scatter plot
 * This plots points on the x- y axis to show relationships btwn two variables
 Best for:
  * Identifying correlation,clusters or outliers
  * Showing how two variables are related i.e weight vs height

5. Histogram
* Groups togethwr continous data into bins to show frequency distribution  
Best for:
 * Visualizing the distribution of a single continous variable
 * Spotting patterns like skewness and data spread.

6. Heatmap
* Uses color to represent values in a grid or matrix format.
Best for:
 * Displaying data density , concentration and intensity
 * Analyzing patterns accross categories and time, often in matrix form

7. Box plots (Box and whisker plot)
* Shows data distribution through quartiles, highlighting median and outliers.

8. Area chart
* Similar to a line chart but with the area btwn the line and the x axis filled in


Best for:
 * Showing cumulatinbe trends over time
 * Visualizing part to whole relationships over time

9. Bubble chart
* Extension of the sacatter plot adding a third dimension with bubble size.




# Line graph

```python
import matplotlib.pyplot as plt

x = [5, 10, 15, 20, 25]
y1 = [3, 9, 27, 18, 15]
y2 = [3, 2, 15, 6, 9]

plt.plot(x, y1, label="A") # Line chart
plt.plot(x, y2, label="B") # Line chart
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart")
plt.legend()
plt.show()
```

Annotations - They can be ussed to add text, arrows or other forms to a plot. They are used to draw attention to key areas or elements in a plot.
* first parameter is the text or the word you want to be displayer on the chart or visualization
* The second parameter is xy, which denotes the place you need to highlight in the graph.
* The third parameter is xytext, which shows where the word should be placed.
* arrowpops facecolor for the color of the arrow and shrink is the distance btwn the arrow point and the line point.

```python
x = [1, 2, 3, 4, 5]
y1 = [3, 9, 27, 18, 15]

plt.plot(x, y1, label="Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart")

plt.annotate("High", xy=(3, 27), xytext=(3, 9), arrowprops=dict(facecolor="green", shrink=0.05))
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [3, 9, 27, 18, 15]

plt.plot(x, y1, label="Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart")

plt.annotate("High", xy=(3, 27), xytext=(3, 9), arrowprops=dict(facecolor="green", shrink=0.05))
plt.annotate("Low", xy=(1, 3), xytext=(1, 20), arrowprops=dict(facecolor="blue", shrink=-0.05))
plt.show()
```

# Line styles - specifies the appearance of a line plot. It is used to modify the lines appeaance such as its thickness, color and dashes




```python
x = [1, 2, 3, 4, 5]
y1 = [3, 9, 27, 18, 15]
y2 = [2, 4, 6, 8, 10]
y3 = [5, 10, 15, 20, 25]
y4 = [5, 5, 5, 7, 8]
y5 = [10, 5, 10, 3, 16]

# plt.plot(x, y, label="line 1")
plt.plot(x, y1, label="line 2", linestyle="--")
plt.plot(x, y2, label="line 3", linestyle=":")
plt.plot(x, y3, label="line 4", linestyle="-")
plt.plot(x, y4, label="line 5", linestyle="-.")
plt.plot(x, y5, label="line 6", linestyle="dotted")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Customized line graph")
plt.show()

```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]
plt.plot(x, y1, label="line 1")
plt.plot(x, y2, label="line 2", marker="*")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.title("Line plot with a marker")
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]

plt.plot(x, y1, label="line 1")
plt.plot(x, y2, label="line 2", marker="*")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.title("Line plot with a marker")

plt.legend()

plt.xlim(1.8, 2.2)
plt.ylim(2, 6)

plt.show()
```

Setting grid background - the grid is a collection of lines drawn behind the data points in a plot, you can change the linestyle to solid, dashed, dotted, dashdot and "". Then set alpha for the transparency value.

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]

plt.plot(x, y1, label="line 1")
plt.plot(x, y2, label="line 2", marker="*")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.title("Line plot with a marker")

plt.legend()

plt.grid(True, linestyle="solid", alpha=0.9, linewidth=1)#setting grid

plt.show()
```

Setting the background color

The set_facecolor() can be used to change the background color

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]

plt.plot(x, y1, label="line 1", color="green")
plt.plot(x, y2, label="line 2", color="red", marker="*")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.title("Line plot with a marker")

plt.legend()

plt.gca().set_facecolor("#f0f0f0")#setting background color

plt.show()

plt.show()
```
